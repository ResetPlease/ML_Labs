4. Напишите функцию, которая принимает 3 трехмерных вектора $x, y, z$ и вычисляет объем пирамиды, образованной этими векторами (используйте смешанное произведение векторов)

```python
def volume(x : np.array, y : np.array, z : np.array) -> int :
    return abs(np.dot(np.cross(x,y),z))
v1 = np.random.randint(1,10,3)
v2 = np.random.randint(1,10,3)
v3 = np.random.randint(1,10,3)
print(f'v1:{v1}\nv2:{v2}\nv3:{v3}')
volume(v1,v2,v3)
```

5. Напишите функцию, которая принимает число N и строит матрицу NxN, с нулями внутри и единицами на границе.

```python
def make_quadro(N: int) -> np.array :
    result = np.ones((N,N))
    result[1:N-1,1:N-1] = np.zeros((N-2,N-2))
    return result
make_quadro(10)
```

10. Даны 2 массива Х и Y, постройте матрицу Коши $C$ ($C_{ij} = \frac{1}{x_i - y_j}$)

```python
def make_koshi(X: np.array, Y: np.array) -> np.array :
    a = len(X)
    b = len(Y)
    C = np.zeros((len(X), len(Y)))
    for i in range(a):
        for j in range(b):
            C[i][j] = X[i] / Y[j]
    return 1/C

X = np.random.rand(10)
Y = np.random.rand(10)

result = make_koshi(X,Y)
print(result)
result.shape
```

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Сделайте круговую диаграмму, на которой были бы отражено число заболеваний COVID-19 по континентам

```python
import matplotlib.pyplot as plt

fig, plot = plt.subplots()
graph_data = table.groupby('continent')
labels = graph_data.groups.keys()
Y = graph_data.sum()['total_deaths']

# тут какой-то мусор, чистить не стал на всяк
plot.pie(Y, labels=labels)
plt.show()

print(labels)
print(Y)
```

2. Постройте диаграмму, на которой в порядке убывания были бы отмечены число тестов на COVID-19 в России, Китае, США, Индии и Нигерии

```python
countries = ['Russia', 'China', 'USA', 'India', 'Nigeria']

filtered_table = table[table['location'].isin(countries)]
agg_data = filtered_table.groupby('location')['total_tests'].sum()

agg_data = agg_data.sort_values(ascending=False)

Names = agg_data.index
Values = agg_data.values

plt.figure(figsize=(10, 5))
plt.bar(Names, Values, color='blue', width=0.3)

plt.xlabel('Страны')
plt.ylabel('Общее количество тестов')
plt.title('Суммарное количество тестов по странам')
plt.show()
```

3. Визуализируйте зависимость ВВП страны и случаев заболевания COVID-19

```python
new_table = table.sort_values('gdp_per_capita')
new_table = new_table[ (new_table['gdp_per_capita'].notnull()) & (new_table['total_cases'].notnull()) ]

gdp = new_table['gdp_per_capita']
covid_cases = new_table['total_cases']


fig, plot = plt.subplots()
plot.plot(gdp, covid_cases)

plot.set(xlabel='gdp per capita', ylabel='total cases',
       title='зависимость ВВП страны и случаев заболевания COVID-19')

plot.grid()
plt.show()
```
